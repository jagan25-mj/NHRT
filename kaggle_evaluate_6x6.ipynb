```python
import os
import glob

# In Kaggle, attached notebook outputs usually appear in /kaggle/input/
# Let's search for the highest step checkpoint in /kaggle/input/
checkpoint_files = glob.glob('/kaggle/input/**/step_*', recursive=True)
checkpoint_files = [f for f in checkpoint_files if 'all_preds' not in f]

if not checkpoint_files:
    print('No checkpoints found! Make sure you attached the output of the previous notebook.')
else:
    # Sort by step number
    def extract_step(f):
        basename = os.path.basename(f)
        return int(basename.replace('step_', ''))
    
    latest_checkpoint = max(checkpoint_files, key=extract_step)
    print(f'Found checkpoint: {latest_checkpoint}')
    
    # Run evaluation
    cmd = f'python evaluate.py checkpoint="{latest_checkpoint}"'
    print(f'Running: {cmd}')
    os.system(cmd)
```
