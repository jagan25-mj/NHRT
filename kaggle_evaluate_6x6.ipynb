```python
import os
import glob

# Let's see what is inside /kaggle/input/ to debug!
print('Contents of /kaggle/input:')
os.system('find /kaggle/input -maxdepth 5')

checkpoint_files = glob.glob('/kaggle/input/**/step_*', recursive=True)
checkpoint_files = [f for f in checkpoint_files if 'all_preds' not in f]

if not checkpoint_files:
    print('\nNo checkpoints found! Make sure you attached the output of the previous notebook.')
    print('If it was an interactive session, you might need to run the training in a "Save Version -> Run All" commit to save the outputs permanently.')
else:
    def extract_step(f):
        basename = os.path.basename(f)
        return int(basename.replace('step_', ''))
    
    latest_checkpoint = max(checkpoint_files, key=extract_step)
    print(f'\nFound checkpoint: {latest_checkpoint}')
    
    cmd = f'python evaluate.py checkpoint="{latest_checkpoint}"'
    print(f'Running: {cmd}')
    os.system(cmd)
```
